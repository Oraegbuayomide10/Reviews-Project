```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# using the pandas function to read the csv file
df = pd.read_csv('C:\\Users\\Leritmwa Danladi\\Downloads\\Scrapped First\\Lagos Hotels.csv')
```

```python
df
```

```python
#doing a .info() on the datframe to get generall information about it
df.info()
```

```python
# changing the string-type entries of the 'Price' column to float
df['Price'] = pd.to_numeric(df.Price.str.replace('[^\d.]', ''), errors='coerce')
```

```python
# changing the string-type entries of the 'Rating' column to float
df['Rating'] = pd.to_numeric(df.Rating.str.replace('[^\d.]', ''), errors='coerce')
```

```python
# dropping the 'Number' column
df.drop('Number', axis = 1, inplace = True)
```

```python
# dropping the 'Like' column
df.drop('Like', axis = 1, inplace = True)
```

```python
# geting the mean of the 'Rating'
mean_rating = round(df['Rating'].mean(),1)
```

```python
# filling the missing values of the "Rating" column with the mean
df['Rating'] = df['Rating'].fillna(mean_rating)
```

```python
# dropping rows with missing values
df.dropna(how = 'any', axis = 0, inplace = True)
```
