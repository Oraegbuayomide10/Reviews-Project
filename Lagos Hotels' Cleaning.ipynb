```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# using the pandas function to read the csv file
df = pd.read_csv('C:\\Users\\Leritmwa Danladi\\Downloads\\Scrapped First\\Lagos Hotels.csv')
```

```python
df
```

```python
#doing a .info() on the datframe to get generall information about it
df.info()
```

```python
# changing the string-type entries of the 'Price' column to float
df['Price'] = pd.to_numeric(df.Price.str.replace('[^\d.]', ''), errors='coerce')
```

```python
# changing the string-type entries of the 'Rating' column to float
df['Rating'] = pd.to_numeric(df.Rating.str.replace('[^\d.]', ''), errors='coerce')
```

```python
# dropping the 'Number' column
df.drop('Number', axis = 1, inplace = True)
```

```python
# dropping the 'Like' column
df.drop('Like', axis = 1, inplace = True)
```

```python
# geting the mean of the 'Rating'
mean_rating = round(df['Rating'].mean(),1)
```

```python
# filling the missing values of the "Rating" column with the mean
df['Rating'] = df['Rating'].fillna(mean_rating)
```

```python
# dropping rows with missing values
df.dropna(how = 'any', axis = 0, inplace = True)
```

```python
df
```

```python
df.to_csv('Lagos Hotels Cleaned.csv')
```

```python
df['City'].value_counts().plot.bar(figsize = (20, 10)) 
```

```python
df['Cleaned'] = df["Review_Text"].apply(lambda x: x.lower())
```

```python
import string
import re
```

```python
# removing punctuations
df['Cleaned'] = df['Cleaned'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
```

```python
# removing the spaces caused bby the removal of punctations
df['Cleaned']=df['Cleaned'].apply(lambda x: re.sub(' +',' ',x))
```

```python
for index,text in enumerate(df['Cleaned'][35:40]):
    print('Review %d:\n'%(index+1),text)
```

```python
from wordcloud import WordCloud
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
```

```python
df['lemmatized']=df['Cleaned'].apply(lambda x: ' '.join([token.lemma_ for token in list(nlp(x)) if (token.is_stop==False)]))
```

```python
df_grouped=df[['Hotel_Title','lemmatized']].groupby(by='Hotel_Title').agg(lambda x:' '.join(x))
df_grouped.head()
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                colormap="Dark2",  
                min_font_size = 10).generate(' '.join(df['lemmatized'])) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
