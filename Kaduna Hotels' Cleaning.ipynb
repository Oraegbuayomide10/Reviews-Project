```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#using the pandas function to read the csv file
df_1 = pd.read_csv("C:\\Users\\Leritmwa Danladi\\Downloads\\Scrapped First\\Kaduna Hotels.csv")
```

```python
#doing a .info() on the datframe to get generall information about it
df_1.info()
```

```python
# changing the string-type entries of the 'Price' column to numeric values
df_1['Price'] = pd.to_numeric(df_1.Price.str.replace('[^\d.]', ''), errors='coerce')
```

```python
# changing the string-type entries of the 'Rating' column to numeric values
df_1['Rating'] = pd.to_numeric(df_1.Rating.str.replace('[^\d.]', ''), errors='coerce')
```

```python
#dropping the 'Number of Reviews' column
df_1.drop('Number_of_Reviews', axis = 1, inplace = True)
```

```python
#geting the mean of the 'Rating' column and rounding it off to one decimal place
mean = round(df_1['Rating'].mean(),1)
```

```python
mean
```

```python
#fiiling the null values of the 'Rating' column with the overall mean of the column
df_1['Rating'] =df_1['Rating'].fillna(mean)
```

```python
df_1
```

```python
#dropping rows with any NaNs in them
df_1.dropna(how = 'any', axis = 0, inplace = True)
```

```python
# saving the cleaned dataset as csv
df_1.to_csv('Kaduna Hotels Cleaned.csv')
```

```python
df_1['Cleaned'] = df_1["Review_Text"].apply(lambda x: x.lower())
```

```python
import string
import re
```

```python
# removing punctuations
df_1['Cleaned'] = df_1['Cleaned'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
```

```python
# removing the spaces caused bby the removal of punctations
df_1['Cleaned']=df_1['Cleaned'].apply(lambda x: re.sub(' +',' ',x))
```

```python
for index,text in enumerate(df_1['Cleaned'][35:40]):
    print('Review %d:\n'%(index+1),text)
```

```python
from wordcloud import WordCloud, STOPWORDS
import spacy
```

```python
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
```

```python
df_1['lemmatized']=df_1['Cleaned'].apply(lambda x: ' '.join([token.lemma_ for token in list(nlp(x)) if (token.is_stop==False)]))
```

```python
df_1_grouped=df_1[['Hotel_Title','lemmatized']].groupby(by='Hotel_Title').agg(lambda x:' '.join(x))
df_1_grouped.head()
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                colormap="Dark2",  
                min_font_size = 10).generate(' '.join(df_1['lemmatized'])) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
